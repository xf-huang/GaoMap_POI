```python
# 读取csv文件
df_gc = pd.read_excel("工厂1.xlsx")
df_jz = pd.read_excel("建筑公司1.xlsx")
df_jxdz = pd.read_excel("机械电子1.xlsx")
df_yy = pd.read_excel("医药1.xlsx")
df_yjhg = pd.read_excel("冶金化工1.xlsx")
df_nlmy = pd.read_excel("农林牧渔基地1.xlsx")
```

```python
df_gc["type"] = "gongchang"
```
